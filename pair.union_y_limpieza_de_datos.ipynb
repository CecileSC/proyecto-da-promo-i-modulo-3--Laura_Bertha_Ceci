```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
# Unión: Asegúrate de realizar los siguientes pasos:

# Explora y carga ambos conjuntos de datos en pandas DataFrames.
df1= pd.read_csv('world-data-2023_part1.csv', index_col=0)
df1
```

```python
df2= pd.read_csv('world-data-2023_part2.csv', index_col=0)
df2
```

```python
# Identifica las columnas comunes entre los dos conjuntos de datos.
df1.columns
print(df1.columns)
df2.columns
print(df2.columns)
```

```python
# Utiliza el método de unión de Pandas que consideres más adecuado para combinar los datos de ambos años en un solo DataFrame.
df3= df1.merge(df2, left_on='Country', right_on='country')
df3
```

```python
# Explica por qué elegiste ese método de unión y cómo se llevaron a cabo los pasos anteriores.
# lo hemos elegido porque tienen una columna en comun llamada 'country'. 
#pasos: leer los dos .csv, ver las columnas de cada uno y elegir un metodo de union (merge)
```

```python
# Ejercicios de Limpieza
# 1-Después de la unión de datos, tenemos dos columnas de "country". Elimina una de ellas.

df3.drop('country',axis = 1, inplace= True)
df3
```

```python
# 2-Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:

# --No tengan espacios.
df3_columnas_nuevas = {columna:columna.replace(' ', '') for columna in df3.columns}
df3_columnas_nuevas

```

```python
# --Estén en minúscula.
df3_columnas_nuevas = {columna:columna.lower().replace(' ', '').replace('(km2)', '').replace('%', '').replace('\n', '').replace(':', '')  for columna in df3.columns}
df3_columnas_nuevas


# --No tengan paréntesis, es decir, quitar "(%)", "(Km2)".ok

# --Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.ok

# --Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.ok
```

```python
df3.rename(columns = df3_columnas_nuevas, inplace=True)
df3
```

```python
#3-La columnas coordinates tiene la latitud y la longitud en una sola columna. 
# Crea dos columnas nuevas, una con la longitud y otra con la latitud. 
# Una vez hecho, elimina la columna de coordinates.
df3[['latitud', 'longitude']] = df3['coordinates'].str.split(',', expand= True)
df3
df3.drop('coordinates',axis = 1, inplace= True)
df3
```
